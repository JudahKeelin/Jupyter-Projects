```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load pruned dataset ---
df = pd.read_csv("features_ravdess_pruned.csv")

# --- Separate features and labels ---
X = df.drop(columns=["label", "filename"], errors='ignore')
y = df["label"]

# --- Encode target labels ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# --- Train-test split ---
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# --- MLP pipeline ---
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("mlp", MLPClassifier(max_iter=1200, random_state=42))
])

# --- GridSearchCV for MLP ---
param_grid = {
    "mlp__hidden_layer_sizes": [(100,), (150,), (100, 50)],
    "mlp__activation": ["relu", "tanh"],
    "mlp__alpha": [0.0001, 0.001],
    "mlp__learning_rate": ["constant", "adaptive"]
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)

# --- Evaluate best model ---
best_mlp = grid_search.best_estimator_
y_pred = best_mlp.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))

# --- Confusion matrix ---
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("MLP (Pruned Features) Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()
```

```python
import joblib

# Save the entire trained pipeline (scaler + MLP + label encoder if needed)
joblib.dump(best_mlp, "mlp_pruned_model.joblib")
print(" Model saved as mlp_pruned_model.joblib")
joblib.dump(le, "label_encoder.joblib")
```
