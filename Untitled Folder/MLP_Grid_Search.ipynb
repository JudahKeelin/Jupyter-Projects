```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df = pd.read_csv("features_ravdess_05.csv")

# Load the features
X = df.drop(columns=["filename", "label"])
y = df["label"]

# Encode emotion labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)
```

```python
# MLP pipeline (with scaler)
mlp_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("mlp", MLPClassifier(max_iter=500, random_state=42))
])

# Define hyperparameter grid
param_grid = {
    "mlp__hidden_layer_sizes": [(100,), (128, 64), (256, 128, 64)],
    "mlp__activation": ["relu", "tanh"],
    "mlp__alpha": [0.0001, 0.001],  # L2 regularization strength
}
```

```python
grid_search = GridSearchCV(
    estimator=mlp_pipe,
    param_grid=param_grid,
    scoring="accuracy",
    cv=5,
    verbose=2,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

# View best parameters
print("Best Parameters:", grid_search.best_params_)
print("Best CV Accuracy:", grid_search.best_score_)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

best_mlp = grid_search.best_estimator_
y_pred = best_mlp.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("MLP (Tuned) Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
from sklearn.inspection import permutation_importance

# Run permutation importance on your full pipeline
result = permutation_importance(best_mlp, X_test, y_test, n_repeats=10, random_state=42)

# Extract and sort
perm_df = pd.DataFrame({
    "feature": X.columns,
    "importance": result.importances_mean
}).sort_values(by="importance", ascending=False)

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.barh(perm_df["feature"].head(20), perm_df["importance"].head(20))
plt.gca().invert_yaxis()
plt.title("Top 20 Feature Importances (Permutation - MLP)")
plt.xlabel("Mean Decrease in Accuracy")
plt.show()
```

```python
from sklearn.inspection import permutation_importance
import pandas as pd
import matplotlib.pyplot as plt

# Run permutation importance on your best pipeline
result = permutation_importance(best_mlp, X_test, y_test, n_repeats=10, random_state=42)

# Combine into a DataFrame
perm_df = pd.DataFrame({
    "feature": X.columns,
    "mean_importance": result.importances_mean,
    "std_importance": result.importances_std
}).sort_values(by="mean_importance", ascending=False)

# Display the full table
import IPython.display as display
display.display(perm_df)

# Optional: Save to CSV
perm_df.to_csv("mlp_permutation_feature_importance.csv", index=False)

# Optional: Plot all instead of top 20
plt.figure(figsize=(12, len(perm_df) * 0.3))
plt.barh(perm_df["feature"], perm_df["mean_importance"])
plt.gca().invert_yaxis()
plt.xlabel("Mean Decrease in Accuracy")
plt.title("All Feature Importances (Permutation - MLP)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load full dataset with actual feature values
features_df = pd.read_csv("features_ravdess_05.csv")

# Load permutation importance rankings
importance_df = pd.read_csv("mlp_permutation_feature_importance.csv")

# Set threshold for pruning
threshold = 0.0025
important_features = importance_df[importance_df["mean_importance"] >= threshold]["feature"].tolist()

# Keep essential metadata columns if present
always_keep = ["label", "filename"]
final_columns = [col for col in important_features if col in features_df.columns] + [
    col for col in always_keep if col in features_df.columns
]

# Prune dataset
pruned_df = features_df[final_columns]

# Save pruned dataset with training data
pruned_df.to_csv("features_ravdess_pruned.csv", index=False)
print(f"✅ Saved pruned dataset with shape {pruned_df.shape} to 'features_ravdess_pruned.csv'")

```
