```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# === Load your full dataset ===
df = pd.read_csv("combined_features.csv")

# === Encode labels ===
le = LabelEncoder()
df['emotion_encoded'] = le.fit_transform(df['emotion'])

# === Feature matrix and label ===
X = df.drop(columns=['emotion', 'emotion_encoded'])
y = df['emotion_encoded']
emotion_labels = le.classes_

# === Train-test split ===
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# === Standardize ===
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# === 1. General model: Random Forest ===
general_model = RandomForestClassifier(n_estimators=100, random_state=42)
general_model.fit(X_train_scaled, y_train)
general_probs = general_model.predict_proba(X_test_scaled)
general_preds = general_model.predict(X_test_scaled)

# === 2. Binary classifier: happy vs angry ===
# Get class indices
happy_idx = list(le.classes_).index('happy')
angry_idx = list(le.classes_).index('angry')

binary_df = df[df['emotion_encoded'].isin([happy_idx, angry_idx])]
X_bin = binary_df.drop(columns=['emotion', 'emotion_encoded'])
y_bin = binary_df['emotion_encoded']

X_bin_train, X_bin_test, y_bin_train, y_bin_test = train_test_split(X_bin, y_bin, stratify=y_bin, test_size=0.2, random_state=42)

scaler_bin = StandardScaler()
X_bin_train_scaled = scaler_bin.fit_transform(X_bin_train)
X_bin_test_scaled = scaler_bin.transform(X_bin_test)

binary_model = RandomForestClassifier(n_estimators=100, random_state=42)
binary_model.fit(X_bin_train_scaled, y_bin_train)

# === 3. Hybrid Prediction Logic ===
hybrid_preds = []
confidence_threshold = 0.65

for i, probs in enumerate(general_probs):
    max_prob = np.max(probs)
    general_pred = general_preds[i]

    if general_pred in [happy_idx, angry_idx] and max_prob < confidence_threshold:
        # fallback to binary model
        binary_probs = binary_model.predict_proba(scaler_bin.transform([X_test.iloc[i]]))[0]
        hybrid_pred = happy_idx if binary_probs[0] > binary_probs[1] else angry_idx
    else:
        hybrid_pred = general_pred

    hybrid_preds.append(hybrid_pred)

# === Evaluation ===
print("=== Classification Report (Hybrid) ===")
print(classification_report(y_test, hybrid_preds, target_names=emotion_labels))

# === Confusion Matrix ===
conf_matrix = confusion_matrix(y_test, hybrid_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=emotion_labels, yticklabels=emotion_labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix - Hybrid Model")
plt.tight_layout()
plt.show()
```
